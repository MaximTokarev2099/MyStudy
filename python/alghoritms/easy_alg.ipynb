## Задача 1: Средний элемент

Рассмотрим три числа a, b и c. Упорядочим их по возрастанию.

Какое число будет стоять между двумя другими?

**Формат ввода**

В единственной строке записаны три целых числа a, b и c (−1000≤a,b,c≤1000), числа разделены одиночными пробелами.

**Формат вывода**

Выведите число, которое будет стоять между двумя другими после упорядочивания.

**Задача с CodeRun:** https://coderun.yandex.ru/problem/median-out-of-three?compiler=python

```python
import sys
from random import randint

def middle_value(a, b, c): # Корявая реализация сортировки пузырьком с нахождением среднего значения
    data = [a, b, c]
    
    i = 0
    count = 0
    while count < 100:
        if i == 2:
            i = 0
        
        if data[i] > data[i + 1]:
            data[i], data[i + 1] = data[i + 1], data[i]
        
        count += 1
        i+=1
    return data[1]


args = [(randint(-1000, 1000), randint(-1000, 1000), randint(-1000, 1000)) for i in range(10)] # Генерируем тестовый набор данных

for i in range(10):
    print(f"Для набора данных {args[i]} среднее значение равно {middle_value(args[i][0], args[i][1], args[i][2])}")
```
